```python
import pandas as pd
df = pd.read_csv('student_performance.csv')
```

```python
print("===================================")
print("STUDENT PERFORMACE EXPLORER REPORT")
print("===================================")

print("\n[1] DATASET OVERVIEW")
print("-----------------------------------")
print("Total students: ",df.shape[0])
print("Total columns: ",df.shape[1])
print("Departments: ",len(df['department'].unique()))

print("\n[2] DEPARTMENT-WISE STUDENT COUNT")
print("-----------------------------------")
print("Department")
print(df['department'].value_counts())

print("\n[3] SCORE ANALYSIS")
print("-----------------------------------")
print("Highest Math Score: ",df['math_score'].max())
print("Lowerst Math Score: ",df['math_score'].min())
print("Average Math Score: ",df['math_score'].mean().round(2))
print("Average Programming Score: ",df['programming_score'].mean().round(2))
print("Average Attendance Percentage Score: ",df['attendance_percentage'].mean().round(2))


print("\n[4] GENDER-WISE AVERAGE SCORES")
print("-----------------------------------")
df['total_score'] = df['math_score'] + df['science_score'] + df['english_score'] + df['programming_score']
gender_scores = df.groupby('gender')['total_score'].mean().round(2)
print(gender_scores)

print("\n[5] TOP 5 PERFORMERS")
print("-----------------------------------")
top5 = df.sort_values('total_score',ascending=False).head(5)
for index , row in top5.iterrows():
  print(f"{row['name']:<20} | {row['department']: <18} | Score: {row['total_score']}")

print("\n[6] STUDENTS WITH LOW ATTENDANCE(Below 75%)")
print("-----------------------------------")
low_att = df[df['attendance_percentage'] < 75][['name','department','attendance_percentage']]
if len(low_att)>0:
  print(low_att.to_string(index=False))
else:
  print("All studends have attendance above 75%")


print("\n===================================")
print("REPORT COMPLETE")
print("===================================")
```
